***Adding products by checking if it exists already***

```python
prod_id = str (input("Enter the product id:"))
name = str (input("Enter the name: "))
qn = str (input("Enter the quantity:"))
pr = str (input("Enter the price:"))
Mfg = str (input("Enter the mfg date:"))
Exp = str (input("Enter the exp date:"))

if  (prod_id in record.keys()):
   print("product exist")
   record[prod_id]['qn'] = record[prod_id]['qn'] + int(qn)
else:
   print("product does not exist")
   record[prod_id] = {"name":name,"qn":qn,"pr":pr,"Mfg":Mfg,"Exp":Exp}

   js = json.dumps(record)

   fd = open ("record.json",'w')
   fd.write(js)
   fd.close()
```
